```python
import pandas as pd
import numpy as np
import sys
from numpy import concatenate
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.semi_supervised import LabelPropagation
from sklearn.linear_model import LogisticRegression
from sklearn.feature_extraction.text import CountVectorizer
```

```python
# define dataset
dataset = pd.read_csv('data_IndosatCare.csv', delimiter=",")
# dataset.info()

y = np.array(dataset.pop('label'))
X = np.array(dataset.pop('text'))
X, y
```

```python
# split into train and test
sentences_train, sentences_test, y_train, y_test = train_test_split(X, y, train_size=299, shuffle=False)

sentences_train, y_train
```

```python
# create a feature vector of the count of the words
vectorizer = CountVectorizer()
vectorizer.fit(sentences_train)

X_train = vectorizer.transform(sentences_train)
X_test  = vectorizer.transform(sentences_test)

X_train
```

```python
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
y_result = classifier.predict(X_test)

y_result
```

```python
dataset = pd.read_csv('data_IndosatCare.csv', delimiter=",")

i = 0
for x in range (299, len(dataset.text)):
  dataset.at[x, 'label'] = y_result[i]
  i=i+1

dataset.to_csv('indosatcare-labelled.csv', index = False)
```
